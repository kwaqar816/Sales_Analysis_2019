### Now we will access the initial two words or Order Date

```python
all_data['Month'] = all_data['Order Date'].str[0:2]
#all_data['Month'] = all_data['Month'].astype('int32')
temp_data = all_data[all_data['Order Date'].str[0:2] == 'Or']
temp_data.head()
```

**Now we have get all data's with or, we will do a little trick here

**Instead of deleting all 'Or' values we will change the search to not equal to instead or equal to

```python
all_data = all_data[all_data['Order Date'].str[0:2] != 'Or']
all_data.head()
```

**Now we have to multiply Quantity Ordered to Price Each for gettig sales data
But we can't do bcz it's in string value first we have to make it int

```python
all_data['Quantity Ordered'] = pd.to_numeric(all_data['Quantity Ordered'])
all_data['Price Each'] = pd.to_numeric(all_data['Price Each'])
```

```python
all_data['Sales'] = all_data['Quantity Ordered'] * all_data['Price Each']
all_data.head()
```

### Here we are taking a city name with state name using apply method lambda function and split() method

```python
def get_city(x):
    return x.split(',')[1]

def get_state(x):
    return x.split(',')[2].split(' ')[1]

all_data['Cities'] = all_data['Purchase Address'].apply(lambda x: get_city(x) + ' ' +get_state(x))

all_data
```

```python
results = all_data.groupby('Cities').sum()
results
```

```python
import matplotlib.pyplot as plt

cities = [city for city,df in all_data.groupby('Cities')]

plt.bar(cities , results['Sales'])
plt.xticks(cities,rotation='vertical')
plt.ylabel('Sales in US Dollors($)')
plt.xlabel('City names')
plt.show()
```

### Which products are sold most

**If two product has same order_id it means that they both are ordered at same time and at same address, meaning a single user baught those two product togetherly

```python
all_data.head()
```

***First we will collect all the duplicate Order ID together and group them

```python
df = all_data[all_data['Order ID'].duplicated(keep=False)]
df.head(10)
```

**Above output prints the first 20 duplicated order id's together

**Now we will create another column where we will store two different products with same order id

```python
df['Grouped'] = df.groupby('Order ID')['Product'].transform(lambda x : ','.join(x))
df.head()
```

***Now we will erase all the duplicates value

```python
df = df[['Order ID' , 'Grouped']].drop_duplicates()
df.head()
```

### Now we will find which products are sold most and to do it we will import two libraries 

```python
from itertools import combinations 
from collections import Counter


count = Counter()

for row in df['Grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list,2)))

    
for key ,value in count.most_common(10):
    print(key,value)

```
